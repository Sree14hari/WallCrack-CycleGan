Add error Columns

```python
import pandas as pd

# --- 1. Define your file paths ---
# Change this to the name of your rounded CSV file (e.g., 'blur_crack_width_comparison_rounded.csv')
input_csv_path = 'shadow.csv'

# This will be the name of the new file with the error columns
output_csv_path = 'shadow_errors.csv'

try:
    # --- 2. Read the CSV file ---
    df = pd.read_csv(input_csv_path)

    # --- 3. Calculate the new error columns ---

    # Check if this is the blur or shadow analysis file and calculate accordingly
    if 'blurred_width_mm' in df.columns:
        # Calculate the error introduced by blurring
        df['Error_Blurred'] = (df['original_width_mm'] - df['blurred_width_mm']).abs()
        # Calculate the final error after CycleGAN restoration
        df['Error_CycleGAN'] = (df['original_width_mm'] - df['restored_width_mm']).abs()
        print("Calculated errors for Blur analysis.")

    elif 'shadowed_width_mm' in df.columns:
        # Calculate the error introduced by the shadow
        df['Error_Shadowed'] = (df['original_width_mm'] - df['shadowed_width_mm']).abs()
        # Calculate the final error after CycleGAN restoration
        df['Error_CycleGAN'] = (df['original_width_mm'] - df['restored_width_mm']).abs()
        print("Calculated errors for Shadow analysis.")
    else:
        print("Error: Could not find 'blurred_width_mm' or 'shadowed_width_mm' columns.")
        # Exit if columns not found
        exit()

    # Round the new error columns for consistency
    df = df.round(4)

    # --- 4. Save the new CSV file ---
    df.to_csv(output_csv_path, index=False)

    print(f"\nSuccessfully calculated error columns.")
    print(f"New file saved as: {output_csv_path}")
    print("\nPreview of the final data with error columns:")
    print(df.head())

except FileNotFoundError:
    print(f"Error: The file '{input_csv_path}' was not found. Please make sure the filename is correct.")
except Exception as e:
    print(f"An error occurred: {e}")
```

Graphs

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSVs
blur_df = pd.read_csv('blur_errors.csv')
shadow_df = pd.read_csv('shadow_errors.csv')

# --- 1. Descriptive statistics ---
print("===== Blur Dataset Statistics =====")
print(blur_df.describe())
print("\n===== Shadow Dataset Statistics =====")
print(shadow_df.describe())

# --- 2. Mean & Variance ---
print("\n===== Blur Means =====")
print(blur_df.mean(numeric_only=True))
print("\n===== Blur Variance =====")
print(blur_df.var(numeric_only=True))

print("\n===== Shadow Means =====")
print(shadow_df.mean(numeric_only=True))
print("\n===== Shadow Variance =====")
print(shadow_df.var(numeric_only=True))

# --- 3. Correlation Heatmaps ---
plt.figure(figsize=(8,6))
sns.heatmap(blur_df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Blur Dataset Correlation Heatmap")
plt.show()

plt.figure(figsize=(8,6))
sns.heatmap(shadow_df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Shadow Dataset Correlation Heatmap")
plt.show()

# --- 4. Error Distribution ---
plt.figure(figsize=(10,5))
sns.histplot(blur_df['Error_Blurred'], color="red", kde=True, label="Error Blurred")
sns.histplot(blur_df['Error_CycleGAN'], color="blue", kde=True, label="Error CycleGAN")
plt.legend()
plt.title("Blur Dataset Error Distribution")
plt.show()

plt.figure(figsize=(10,5))
sns.histplot(shadow_df['Error_Shadowed'], color="orange", kde=True, label="Error Shadowed")
sns.histplot(shadow_df['Error_CycleGAN'], color="green", kde=True, label="Error CycleGAN")
plt.legend()
plt.title("Shadow Dataset Error Distribution")
plt.show()

# --- 5. Boxplots for Comparison ---
plt.figure(figsize=(8,6))
sns.boxplot(data=blur_df[['Error_Blurred','Error_CycleGAN']])
plt.title("Blur Dataset Error Comparison")
plt.show()

plt.figure(figsize=(8,6))
sns.boxplot(data=shadow_df[['Error_Shadowed','Error_CycleGAN']])
plt.title("Shadow Dataset Error Comparison")
plt.show()

# --- 6. Scatter: Clean vs Restored ---
plt.figure(figsize=(6,6))
plt.scatter(blur_df['original_width_mm'], blur_df['restored_width_mm'], alpha=0.6)
plt.plot([0,10],[0,10], 'r--')
plt.xlabel("Clean Width (mm)")
plt.ylabel("Restored Width (mm)")
plt.title("Blur Dataset: Clean vs Restored")
plt.show()

plt.figure(figsize=(6,6))
plt.scatter(shadow_df['original_width_mm'], shadow_df['restored_width_mm'], alpha=0.6)
plt.plot([0,10],[0,10], 'r--')
plt.xlabel("Clean Width (mm)")
plt.ylabel("Restored Width (mm)")
plt.title("Shadow Dataset: Clean vs Restored")
plt.show()
```
